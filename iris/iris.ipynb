```python
# 导入数据可视化模块
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

%matplotlib inline
plt.style.use({"figure.figsize":(10,8)})

sns.set_style("whitegrid")
```

```python
# load iris dataset from net
# data = datasets.load_iris()
# irisDataFrame = pd.DataFrame(data.data, columns=data.feature_names)
# load iris from local
irisDataFrame = sns.load_dataset('iris',data_home='~/datasets/datasets',cache=True, index_col=0)
irisDataFrame
```

## 数据内容

- sepal_length: 花萼长度
- sepal_widht: 花萼宽度
- petal_length: 花瓣长度
- petal_width: 花瓣宽度
- species: 鸢尾花所属的亚种

```python
# 取出所有为setosa的鸢尾花数据
setosa = irisDataFrame.loc[irisDataFrame.Species == 'setosa']

# 取出所有为virginica的鸢尾花数据
virginica = irisDataFrame.loc[irisDataFrame.Species == 'virginica']

# 取出所有为versicolor的鸢尾花数据
versicolor = irisDataFrame.loc[irisDataFrame.Species == 'versicolor']

# 分别绘制setosa和virginica两种鸢尾花的花萼宽度，花萼长度两个维度的KDE可视化
ax = sns.kdeplot(setosa['Sepal.Width'], setosa['Sepal.Length'], cmap="Reds", shade=True, shade_lowest=False)
ax = sns.kdeplot(virginica['Sepal.Width'], virginica['Sepal.Length'], cmap="Blues", shade=True, shade_lowest=False)
# ax = sns.kdeplot(versicolor['Sepal.Width'], versicolor['Sepal.Length'], cmap="Greens", shade=True, shade_lowest=False)
```
